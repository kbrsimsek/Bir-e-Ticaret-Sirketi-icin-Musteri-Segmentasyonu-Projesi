## RFM Analizi ile Müşteri Segmentasyonu

```python
df_new.head()
```

```python
df_new.info()
```

```python
df_new["InvoiceDate"].min()
```

```python
df_new["InvoiceDate"].max()
```

```python
#Bunlar eski veriler olduğu için veri setindeki son tarih bugünün tarihi olarak kullanılacaktır.

import datetime as dt
today_date = dt.datetime(2011,12,9)
```

```python
today_date
```

```python
df_new.groupby("Customer ID").agg({"InvoiceDate":"max"}).head()
```

```python
df_new["Customer ID"] = df_new["Customer ID"].astype(int)
```

```python
(today_date - df_new.groupby("Customer ID").agg({"InvoiceDate":"max"})).head()
```

```python
temp_df = (today_date - df_new.groupby("Customer ID").agg({"InvoiceDate":"max"}))
```

```python
temp_df.rename(columns={"InvoiceDate": "Recency"}, inplace = True)
```

```python
temp_df.head()
```

```python
#İstisnasız saatler

recency_df = temp_df["Recency"].apply(lambda x: x.days)
```

```python
recency_df.head()
```

### Sıklık

```python
temp_df = df_new.groupby(["Customer ID","Invoice"]).agg({"Invoice":"count"})
```

```python
temp_df.head()
```

```python
temp_df.groupby("Customer ID").agg({"Invoice":"count"}).head()
```

```python
freq_df = temp_df.groupby("Customer ID").agg({"Invoice":"count"})
freq_df.rename(columns={"Invoice": "Frequency"}, inplace = True)
freq_df.head()
```

### Parasal

```python
monetary_df = df_new.groupby("Customer ID").agg({"TotalPrice":"sum"})
```

```python
monetary_df.head()
```

```python
monetary_df.rename(columns={"TotalPrice": "Monetary"}, inplace = True)
```

```python
print(recency_df.shape,freq_df.shape,monetary_df.shape)
```

```python
rfm = pd.concat([recency_df, freq_df, monetary_df],  axis=1)
```

```python
rfm.head()
```

```python
#qcut ve sıralama ile 5 çeyreğe bölme
#5 hepsi için en iyisi olacak

rfm["RecencyScore"] = pd.qcut(rfm["Recency"], 5, labels = [5, 4, 3, 2, 1])
```

```python
rfm["FrequencyScore"] = pd.qcut(rfm['Frequency'].rank(method = 'first'), 5, labels = [1, 2, 3, 4, 5])
```

```python
rfm["MonetaryScore"] = pd.qcut(rfm['Monetary'], 5, labels = [1, 2, 3, 4, 5])
```

```python
rfm.head()
```

```python
rfm.describe([0.2,0.3,0.4,0.6,0.8]).T
```

```python
pd.set_option('display.float_format', lambda x: '%.0f' % x)
import matplotlib.pyplot as plt
```

```python
(rfm['RecencyScore'].astype(str) + 
 rfm['FrequencyScore'].astype(str) + 
 rfm['MonetaryScore'].astype(str)).head()
```

```python
rfm["RFM_SCORE"] = rfm['RecencyScore'].astype(str) + rfm['FrequencyScore'].astype(str) + rfm['MonetaryScore'].astype(str)
```

```python
rfm.head()
```

```python
rfm.describe().T
```

```python
# En iyi segment '555'

rfm[rfm["RFM_SCORE"] == "555"].head()
```

```python
rfm[rfm["RFM_SCORE"] == "111"].head()
```

```python
# Normal İfade için (yenilik - sıklık)
seg_map = {
    r'[1-2][1-2]': 'Hibernating',
    r'[1-2][3-4]': 'At Risk',
    r'[1-2]5': 'Can\'t Loose',
    r'3[1-2]': 'About to Sleep',
    r'33': 'Need Attention',
    r'[3-4][4-5]': 'Loyal Customers',
    r'41': 'Promising',
    r'51': 'New Customers',
    r'[4-5][2-3]': 'Potential Loyalists',
    r'5[4-5]': 'Champions'
}
```

```python
rfm['Segment'] = rfm['RecencyScore'].astype(str) + rfm['FrequencyScore'].astype(str)
rfm['Segment'] = rfm['Segment'].replace(seg_map, regex=True)
rfm.head()
```

```python
rfm[["Segment", "Recency","Frequency","Monetary"]].groupby("Segment").agg(["mean","count"])
```

```python
#Bir grupta müşteri bulma

rfm[rfm["Segment"] == "Need Attention"].head()
```

```python
rfm[rfm["Segment"] == "Need Attention"].index
```

```python
# na_df Dikkat Edilmesi Gereken DataFrame
na_df = pd.DataFrame()
na_df["NeedAttentionID"] = rfm[rfm["Segment"] == "Need Attention"].index
```

```python
na_df.head()
```

```python
na_df.to_csv("need_attention.csv")
```

```python
# Her segmentteki müşteri sayısı ve yüzdesi
segments_counts = rfm['Segment'].value_counts().sort_values(ascending=True)

fig, ax = plt.subplots()

bars = ax.barh(range(len(segments_counts)),
              segments_counts,
              color='silver')
ax.set_frame_on(False)
ax.tick_params(left=False,
               bottom=False,
               labelbottom=False)
ax.set_yticks(range(len(segments_counts)))
ax.set_yticklabels(segments_counts.index)

for i, bar in enumerate(bars):
        value = bar.get_width()
        if segments_counts.index[i] in ['At Risk', 'Can\'t Loose', 'About to Sleep', 'Need Attention',]:
            bar.set_color('firebrick')
        ax.text(value,
                bar.get_y() + bar.get_height()/2,
                '{:,} ({:}%)'.format(int(value),
                                   int(value*100/segments_counts.sum())),
                va='center',
                ha='left'
               )

plt.show()
```

##### RFM analizi sonuçlarımıza dayanarak, "risk altında, uyku konusunda, dikkat edilmesi gereken ve kaybedilemez" segmentlerinde yer alan müşterileri elde tutmak için farklı pazarlama teknikleri üzerinde çalışabiliriz.Ayrıca satışlarımızı artırmak için diğer müşterilerimiz üzerinde çalışmamız gerekiyor.
